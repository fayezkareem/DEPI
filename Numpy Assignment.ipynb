```python
mylist=(x*x for x in range(6))
x=np.fromiter(mylist,dtype=int)
print(x)
```

```python
iteration=((x+1,x+2,x+3,x+4) for x in range(6))
x=np.fromiter(iteration,dtype=np.dtype((int,4)))
print(x)
```
